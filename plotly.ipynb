Diagram Sankey'a - mamy jakies zadanie ze mozemy wjakis stanach okreslic strukture naszych klientow np, zagadnienie churn ...churn to taka sytuacja w ktorej klient danej firmy rezygnuje z jej  uslug no i w interesie takiej firmy jest to aby przewidywac w miare mozliwosci tej churn klienta i ewentualnie ofertowac go nowymi promocjami tak aby tego klienta zatrzymac bowiem taniej zatrzymac biezacego klienta niz pozyskac nowegoi czesto pojawiaja potrzeby budowania modeli churn

````
#total = 100
#Non-Churn-2018->Non-churn-2019(65procent)
#Non-Churn-2018->Churn-2019(12procent)
#Churn-2018->Churn-2019(18procent)
#Churn-2018->Non-Churn-2019(5procent)


data1=[go.Sankey(node=dict(label=['Nonchurn_2018', 'Churn_2018','Nonchurn_2019','Churn_2019']),
                 link=dict(source=[0, 0, 1, 1], #indeks odpowiadajacy etykiecie (labels)
                           target=[2, 3, 2, 3],
                           value=[5, 12, 1, 5]))]
fig=go.Figure(data=data1)
fig.show()
````

````
data1=[go.Sankey(node=dict(label=['Nonchurn_2018', 'Churn_2018','Nonchurn_2019','Churn_2019','Nonchurn_2020','Churn_2020']),
                 link=dict(source=[0,0,1,1,2,3], #indeks odpowiadajacy etykiecie (labels)
                           target=[2,3,2,3,4,30],
                           value=[65,10,5,20,70]))]
fig=go.Figure(data=data1)
fig.show()
````

````
def fetch_financial_data(company='AMZN'):
  """
  This function fetch stock market quotations.
  """
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')

df_amzn=fetch_financial_data(company='AMZN')
df_aapl=fetch_financial_data(company='AAPL')
df_msft=fetch_financial_data(company='MSFT')
df_nflx=fetch_financial_data(company='NFLX')
df_twtr=fetch_financial_data(company='TWTR')
````

````
df_amzn.head(2)
````

````
import pandas as pd

df=pd.concat([df_amzn.Close, df_aapl.Close, df_msft.Close, df_nflx.Close, df_twtr.Close], axis=1, join='inner')
df.columns=['amazon', 'apple', 'microsoft', 'netflix', 'twitter']
df=df.reset_index()
df.head()
````

````
df.describe()
````

````
df.info()
````
